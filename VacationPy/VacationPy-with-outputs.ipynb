### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# Heatmap of humidity
locations = weather_data[["Lat", "Lng"]]
humidity = weather_data["Humidity"].astype(float)
```

```python
figure = gmaps.figure(center = [0,0] ,zoom_level = 2)

heat_layer = gmaps.heatmap_layer(locations, weights=humidity,
                               dissipating=False, max_intensity=100,
                               point_radius = 4)
figure.add_layer(heat_layer)
figure
```

```python
hotel_df = ideal_weather_df[["City","Country","Lat","Lng"]]
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
# Set parameters to search for a hotel


# Iterate through 

    # get lat, lng from df

    
    # Use the search term: "Hotel" and our lat/lng
   

    # make request and print url
    
    
    # convert to json
   
    
    # Grab the first hotel from the results and store the name
```

```python
hotels = []

search = "hotel"
target_type = "hotel"
radius = 5000

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
for i in range(len(hotel_df)):
    lat = hotel_df["Lat"]
    lng = hotel_df["Lng"]

    params = {
        "location": f"{lat},{lng}",
        "keyword": search,
        "radius": radius,
        "types": target_type,
        "key": g_key}


    requested = requests.get(base_url, params)
    response = requested.json()
    
    try:
        hotel_df.append(response['results'][0]['name'])
    except:
        hotel_df.append("")
        print(f"Missing field/result... skipping.")
        
        
hotel_df
```
